```python
# Install Packages

!pip install groq

!pip install gradio

# Import the Packages

from google.colab import userdata
from groq import Groq
import gradio

client = Groq(
    api_key=userdata.get("GROQ_API_KEY")
)

# Define a function to give content and role

def initialize_messages():
    return [
        {
            "role": "system",
            "content": """
You are Detective AI, an intelligent virtual detective specializing in solving mysteries,
analyzing evidence, and investigating fictional or user-provided cases.

Your responsibilities include:
- Analyze clues, witness statements, documents, images, and timelines.
- Identify inconsistencies and contradictions.
- Build logical theories based on available evidence.
- Ask follow-up questions when information is missing.
- Distinguish between facts, assumptions, and speculation.
- Present multiple possible explanations when appropriate.
- Explain your reasoning step by step.
- Never claim certainty without sufficient evidence.
- Maintain an objective and unbiased approach.

Rules:
1. Base conclusions only on the evidence provided.
2. Clearly indicate when information is insufficient.
3. Do not fabricate evidence.
4. Ask for additional clues whenever necessary.
5. Organize findings into Facts, Observations, Suspects (if any), Possible Motives, Timeline, and Conclusion.
6. Be professional, logical, and analytical in every response.

Your goal is to help users investigate mysteries by applying deductive reasoning rather than making unsupported accusations.
"""
        }]

# Assign it to a variable

messages_prmt = initialize_messages()

x = [5,7,2]
x.append(33)
print(x)

# Define a function to connect with LLM

def customLLMBot(user_input, history):
    global messages_prmt

    messages_prmt.append({
        "role": "user",
        "content": user_input
    })

    response = client.chat.completions.create(
        model="llama-3.3-70b-versatile",
        messages=messages_prmt,
        temperature=0.3,      # More logical and consistent responses
        max_tokens=1024
    )

    reply = response.choices[0].message.content

    messages_prmt.append({
        "role": "assistant",
        "content": reply
    })

    return reply

# Create an object of chat interface class in gradio

iface = gradio.ChatInterface(
    customLLMBot,
    chatbot=gradio.Chatbot(height=500),
    textbox=gradio.Textbox(
        placeholder="Describe a mystery, crime scene, or upload clues..."
    ),
    title="🕵️ Detective AI",
    description="""
    Solve mysteries using AI-powered investigation.
    Analyze clues, identify suspects, detect contradictions,
    build timelines, and uncover the truth through logical reasoning.
    """,

    examples=[
        "A diamond was stolen from a museum. Three people were present. Who is the most likely suspect?",
        "Analyze this witness statement for contradictions.",
        "Help me build a timeline for a missing person case.",
        "Who is lying based on these three witness statements?",
        "Summarize the evidence collected in this case."
    ]
)

# Call launch function to execute

iface.launch(share=True)
```
